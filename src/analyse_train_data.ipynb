```python
import pandas as pd
import numpy as np
from datetime import datetime

```

```python
# load cleaned train data

df_cleaned = pd.read_csv('https://raw.githubusercontent.com/akwasianing/train_project/refs/heads/main/data_clean/cleaned_train_data.csv')
df_cleaned.head()
```

```python
# creating one more attibute (Car_Age) and adding to dataset

current_year = datetime.now().year
df_cleaned['Car_Age'] = current_year - df_cleaned['Year']
df_cleaned.to_csv('Car_Age_attribute_data.csv', index=False)
df_cleaned.head()
```

```python
#  select specified columns

df_cleaned_selected = df_cleaned[['Name', 'Year', 'Price']]
df_cleaned_selected.to_csv('selected_columns.csv', index=False)
df_cleaned_selected.head()
```

```python
# filter rows where the car price is greater than 35 Lakh

df_cleaned_filtered = df_cleaned[df_cleaned['Price'] > 35].reset_index(drop=True)
df_cleaned_filtered.to_csv('filtered_data.csv', index=False)
df_cleaned_filtered.head()
```

```python
# rename columns: 'Name' → 'Make_and_Model', 'Price' → 'Original_Price'

df_cleaned_renamed = df_cleaned.rename(columns={'Name': 'Make_and Model', 'Price': 'Original_Price'})
df_cleaned_renamed.to_csv('rename.csv', index=False)
df_cleaned_renamed.head()

```

```python
# mutate - create new column for original car price in USD assuming 1 Lakh is 1200 USD

df_cleaned_renamed['Original_Price_USD'] = df_cleaned_renamed['Original_Price'] * 1200
df_cleaned_renamed.to_csv('mutate.csv', index=False)
df_cleaned_renamed.head()
```

```python
# arrange - sort Original_Price in USD in descending order

df_cleaned_sort = df_cleaned_renamed.sort_values(by='Original_Price_USD', ascending=False)
df_cleaned_renamed.to_csv('arrange.csv', index=False)
df_cleaned_sort.head()
```

```python
# summarise with group by operation - Average Price by Location (City)

df_cleaned_renamed.groupby('Location')['Original_Price'].mean().reset_index(name='Avg_Price')
df_cleaned_renamed.to_csv('summarise.csv', index=False)
```

```python
# summarise with group by operation - count of Cars by City

df_cleaned_renamed.groupby(['Location']).size().reset_index(name='Car_Count')
df_cleaned_renamed.to_csv('summarise2.csv', index=False)
```
